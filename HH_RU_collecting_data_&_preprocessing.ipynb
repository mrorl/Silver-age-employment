## Подготовка к сбору вакансий и резюме

Скачаем справочник регионов:

```python
def getRegions():
    """
    Создаем метод для получения страницы со списком регионов.
    """
    req = requests.get('https://api.hh.ru/areas') # Посылаем запрос к API
    data = req.content.decode() # Декодируем его ответ, чтобы Кириллица отображалась корректно
    req.close()
    return data
 
   
# Преобразуем текст ответа запроса в справочник Python
jsObj = json.loads(getRegions())
areas = json_normalize(jsObj)
     
print('Информация о регионах собрана')
```

Скачаем справочник специальностей:

```python
def getSpecializations():
    """
    Создаем метод для получения страницы со списком специализаций.
    """
    req = requests.get('https://api.hh.ru/specializations') # Посылаем запрос к API
    data = req.content.decode() # Декодируем его ответ, чтобы Кириллица отображалась корректно
    req.close()
    return data
 
   
# Преобразуем текст ответа запроса в справочник Python
jsObj = json.loads(getSpecializations())
specializations = json_normalize(jsObj)

     
print('Информация о специализациях собрана')
```

## Сбор вакансий

```python
vacancy_ids = download.vacancy_ids(area_id=area_id, specialization_id=specialization_id,search_period=30, num_pages=20)
```

```python
def getVacancy(id):
    """
    Создаем метод для скачивания вакансий по id.
    """
    vacancy = download.vacancy(id)
    return vacancy

vacancy_list = []

for id in vacancy_ids:
  vacancy_json = getVacancy(id)
  vacancy_list.append(vacancy_json)

df_vacancy = json_normalize(vacancy_list)
df_vacancy.head()
```

```python
len(resume_ids)
```

```python
def getResume(id):
    """
    Создаем метод для скачивания резюме по id.
    """
    resume = download.resume(id)
    resume = parse.resume(resume)
    return resume

resume_list = []

for id in resume_ids:
  resume_json = getResume(id)
  resume_list.append(resume_json)

df_resume = json_normalize(resume_list)
df_resume.head()
```

# Проведение предобработки

```python
import re
import datetime
```

```python
def json_to_list(json_data):
    """
    Создаем метод для преобразования json объекта в список.
    """
    return json_normalize(json_data).values.tolist()


def apply_impl(df):
    """
    Создаем метод для применения функции json_to_list к каждому ряду датафрейма.
    """
    return df.apply(
        lambda row:
          json_to_list(row)
    )
```

Добавление колонки birth_year:

```python
month_dict = {'января': 1, 'февраля': 2,'марта': 3,'апреля': 4,
              'мая': 5,'июня': 6,'июля': 7,'августа': 8,
              'сентября': 9,'октября': 10,'ноября': 11,'декабря': 12,
              'January': 1, 'February': 2,'March': 3,'April': 4,
              'May': 5,'June': 6,'July': 7,'August': 8,
              'September': 9,'October': 10,'November': 11,'December': 12}

def extract_age(birth_date):
    """
    Создаем метод для вычисления возраста по атрибуту birth_date.
    """
    date_month = int(month_dict[birth_date.split()[1]])
    date_year = int(birth_date.split()[2])
    date_day = int(birth_date.split()[0])

    date = datetime.datetime(date_year, date_month, date_day)
    now = datetime.datetime.now()
    age = now - date

    return age.days//365


def apply_extract_age(df):
    """
    Создаем метод для применения функции extract_year к каждому ряду датафрейма.
    """
    return df.apply(
        lambda row:
          extract_age(row)
    )
```

```python
df_resume['age'] = apply_extract_age(df_resume['birth_date'])
df_resume.head()
```
